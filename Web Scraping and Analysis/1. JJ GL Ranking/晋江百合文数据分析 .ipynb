```python
df_g = df.groupby('类型')['风格'].value_counts()
df1 = df_g.unstack(fill_value=0)
a = df1.assign(tmp=df1.sum(axis=1)).sort_values('tmp', ascending=True).drop('tmp', 1)
```

```python
fig = a.plot.barh()
fig.show()
```

```python
df_a = df.groupby('作者').aggregate({'字数':'sum',
                            '作品积分':'sum'})
df_a['字/积分']=df_a['作品积分']/df_a['字数']
```

```python
df_a.sort_values('字/积分',ascending=False).head(10)
```

<div class="alert alert-block alert-info">
<b></b> 请大家记住这些大佬们的名字，因为他们每个字都值上千个积分。(⊙ˍ⊙) 那我们来看一下这些作者大大的作品们。
</div>

```python
df[(df['作者'] == '墨宝非宝') | (df['作者'] == '扶华') | (df['作者'] == '铜香墨臭') | (df['作者'] == '凌睿')| \
   (df['作者'] == '君sola')| (df['作者'] == '一盏夜灯') | (df['作者'] == '妲婴') | \
  (df['作者'] == '叶涩')].sort_values(by = ['作品积分','字数','作者'], ascending=[False,True,True])
```

```python
df_w = df.groupby('作品').aggregate({'字数':'sum',
                            '作品积分':'sum'})
df_w['字/积分_作品']=df_w['作品积分']/df_w['字数']
```

```python
round(df_w.sort_values('字/积分_作品',ascending=False)).head(10)
```
